Analysis for Phosphorous

```python
final_fl_dict['Chemical_Code'] = 19
for code, cid in zip(fl_phosphorous.crop_code, fl_phosphorous.client_id):
    print(code)
    print(cid)
    similar = False
    df = pd.read_sql(f"SET NOCOUNT ON;execute [load_YieldvsSoilTestRanges] @crop_code={code},@Chemical_code=19,@client_id={cid}", con=conn_lims)
    master = pd.read_sql(f"SET NOCOUNT ON;execute [load_YieldvsSoilTestRanges] @crop_code={code},@Chemical_code=19,@client_id=0", con=conn_lims)
    df = df[df.columns.to_list()[11:]]
    master = master[master.columns.to_list()[11:]]
    print(len(df.columns))
    print(len(master.columns))
    if(len(df.columns) != len(master.columns)):
        continue
    for i in df.columns:
        if(i not in master.columns):
            similar=False
            break
        if(df[i].to_list() == master[i].to_list()):
            similar = True
            continue
        else:
            similar = False
            break
    if(similar == True):
        final_fl_dict.loc[final_fl_dict['crop_code'] == code,'SimilarToMaster'] = True
    else:
        final_fl_dict.loc[final_fl_dict['crop_code'] == code,'SimilarToMaster'] = False
```
