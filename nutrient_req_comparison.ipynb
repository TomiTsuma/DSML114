```python
arr = []
```

```python
for client_type,crop_code,crop_name,chemical_code in zip(df['client_type'], df['crop_code'], df['crop_name'],df['chemical_code']):
    _ = df.loc[(df['client_type'] == client_type) & (df['crop_code'] == crop_code) & (df['chemical_code'] == chemical_code )]
    useYieldvsSoilTargetInstances = len(_.loc[_['UsesYieldVsSoilTestRange'] == True])
    yieldvsSoilTargetInstancesMatchesMaster = len(_.loc[_['yieldTargetsMatchesMasterCrop'] == True])
    _Arr = []
    _Arr.append(client_type)
    _Arr.append(crop_code)
    _Arr.append(crop_name)
    _Arr.append(chemical_code)
    _Arr.append(useYieldvsSoilTargetInstances)
    _Arr.append(yieldvsSoilTargetInstancesMatchesMaster)
    arr.append(_Arr)
```

```python
arr = np.array(arr)
```

```python
df_summary = pd.DataFrame(arr, columns = ["client_type","crop_code","crop_name","chemical_code","useYieldvsSoilTargetInstances","yieldvsSoilTargetInstancesMatchesMaster"])
```

```python
df_summary['yieldvsSoilTargetInstancesMatchesMaster'] = df_summary['yieldvsSoilTargetInstancesMatchesMaster'].astype("int")
```
